# Testing the hough transformation

```python
import convert_image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import Hough

# Load header file to array, turn image to 'binary'
edgeImg = convert_image.read_header("camera_man.h", ",")
edgeImg[edgeImg == 255] = 1
```

```python
# call hough transformation function
output_img = Hough.hough_transform(edgeImg)
# Display loaded image, hough space and image with lines
fig = plt.figure(figsize = (12, 4))
fig.add_subplot(1, 2, 1)
plt.imshow(edgeImg, cmap=plt.cm.gray)

fig.add_subplot(1, 2, 2)
plt.imshow(output_img, cmap=plt.cm.gray)
plt.show()
```
